````python
%%file aggregative_news_analyzer.sh
#!/bin/bash

# paths to a folder named datasets and the specific txt files
categories_path="/content/drive/MyDrive/Colab Notebooks/datasets/rcv1-v2.topics.qrels.txt"
terms_path="/content/drive/MyDrive/Colab Notebooks/datasets/lyrl2004_vectors_test_pt0.dat.txt"
stems_path="/content/drive/MyDrive/Colab Notebooks/datasets/stem.termid.idf.map.txt"

# variables and data structures for data
declare -A category_dict
categories_start=73698
categories_end=73800
declare -A terms_dict
terms_lines=50
stems_list=()

# file: Array. Each value is a line of the text file:
#       <category_id> <document_id> 1
#       e.g. GCAT 26153 1
# category_dict: Associative array. Each key is a category and each value is
#                a string with the corresponding documents:
#                e.g. GCAT 26152,26153,26154,26155,26156,
function load_categories() {
    OLDIFS="$IFS"
    IFS=$'\n'
    let num_of_lines="$categories_end"-"$categories_start"
    file=$(head -$categories_end $categories_path | tail -$num_of_lines)
    for line in $file
    do
        category=$(echo $line | cut -f1 -d" ")
        document=$(echo $line | cut -f2 -d" ")
        category_dict[$category]=${category_dict[$category]}$document","
    done
    IFS="$OLDIFS"
}

# file: Array. Each value is a line of the text file:
#       <term_id>  <document_id_1>:<weight_1> <document_id_2>:<weight_2> ...
#       e.g. 26151  1523:0.107308708502892 2293:0.10189882251665 2621:0.234278331371989 ...
# fields: Array like this: [26151, 1523:0.107308708502892, 2293:0.10189882251665, 2621:0.234278331371989, ...]
# document: First value of fields array.
# term: Value like this: 1523
# terms_dict: Associative array. Each key is a term and each value is
#             a string with the corresponding documents:
#             e.g. 26151 1523, 2293, 2621, ...
function load_terms() {
    OLDIFS="$IFS"
    IFS=$'\n'
    file=$(head -$terms_lines $terms_path)
    for line in $file
    do
        IFS=$' '
        fields=($(echo $line))
        document=${fields[0]}
        for ((t=1; t <= ${#fields[@]} - 1; t++))
        do
            term=$(echo ${fields[$t]} | cut -f1 -d":")
            terms_dict[$term]=${terms_dict[$term]}$document","
        done
    done
    IFS="$OLDIFS"
}

function load_stems() {
    stems_list=($(paste <(cut -f1 -d" " "$stems_path")))
}

# $1: num of documents for a term
# $2: num of documents for a category
# $3: num of documents for BOTH a term AND a category (intersection cardinality number)
#
# Returns the score as float with five decimals.
#                                      |intersection|: $3
# J(T, C) = -----------------------------------------------------------------------
#             |docs_per_category|: $2 + |docs_per_term|: $1 - |intersection|: $3
function jaccard_index() {
    let num="$3"
    let denom=$1+$2-$3
    result=$(echo "scale=5; $num/$denom" | bc)
    echo $result
}

# Given a term return stem
# e.g. 40238 -> sunday
function term_to_stem() {
    index="$1"-1
    echo ${stems_list[$index]}
}

# Given a stem return term
# e.g. sunday -> 40238
function stem_to_term() {
    for ((i=0; i <= ${#stems_list[@]} - 1; i++))
    do
        if [[ ${stems_list["$i"]} == "$1" ]]
        then
            let index="$i"+1
            echo $index
            break
        fi
    done
}

# params: $1: category, $2: k
#
# category_docs: Array like this: [ ,doc_id_1, doc_id_2, doc_id_3, doc_id_4, doc_id_5, ...].
#                Contains documents that belong to a specific CATEGORY ($1 argument).
#                NOTE: first element is " " and we ignore it (e.g. let b="${#category_docs[@]}"-1, watch -1)
# m: Variable that takes all the values of the keys of the associative array terms_dict (e.g. m=term_id_1).
# term_docs: Array like this: [ ,doc_id_1, doc_id_2, doc_id_3, doc_id_4, doc_id_5, ...].
#            Contains documents that belong to a specific TERM (m).
#            NOTE: first element is " " and we ignore it (e.g. let a="${#term_docs[@]}"-1, watch -1)
# intersection: Essentially is: |category_docs INTERSECTION term_docs| (cardinality number).
# scores: Return value: string containing <stem1>:<JIScore(stem1, category)> <stem2>:<JIScore(stem2, category)>, ...
#         sorted based on JIScore (decreasing). Has k [$2] stems.
#         e.g. sunday:.88235 world:.60784 win:.48000 (for k = 3).
function command_1 {
    category=$1
    k=$2
    scores=""

    OLDIFS="$IFS"
    IFS=$','
    category_docs=(echo ${category_dict[$category]})
    IFS="$OLDIFS"
    let b="${#category_docs[@]}"-1

    for m in ${!terms_dict[@]}
    do
        OLDIFS="$IFS"
        IFS=$','
        term_docs=(echo ${terms_dict[$m]})
        IFS="$OLDIFS"
        let a="${#term_docs[@]}"-1

        intersection=0
        for ((i=1; i <= ${#term_docs[@]} - 1; i++))
        do
            for ((j=1; j <= ${#category_docs[@]} - 1; j++))
            do
                if [[ ${term_docs[$i]} == ${category_docs[$j]} ]]
                then
                    intersection=`expr $intersection + 1`
                fi
            done
        done

        scores="$scores"$(term_to_stem $m)":"$(jaccard_index "$a" "$b" "$intersection")"\n"
    done

    scores=$(echo -e $scores | sort -t":" -n -r +1)

    echo $(echo $scores | cut -f1-$k -d" ")
}

# params: $1: stem, $2: k
#
# term_docs: Array like this : [ ,doc_id_1, doc_id_2, doc_id_3, doc_id_4, doc_id_5, ...].
#            Contains documents that belong to a specific TERM (m).
#            NOTE: first element is " " and we ignore it (e.g. let a="${#term_docs[@]}"-1, watch -1)
# m: Variable that takes all the values of the keys of the associative array category_dict (e.g. m=category_id_1).
# category_docs: Array like this : [ ,doc_id_1, doc_id_2, doc_id_3, doc_id_4, doc_id_5, ...].
#                Contains documents that belong to a specific CATEGORY ($1 argument).
#                NOTE: first element is " " and we ignore it (e.g. let b="${#category_docs[@]}"-1, watch -1)
# intersection: Essentially is: |category_docs INTERSECTION term_docs| (cardinality number).
# scores: Return value: string containing <category1>:<JIScore(category1, stem)> <category2>:<JIScore(category2, stem)>, ...
#         sorted based on JIScore (decreasing). Has k [$2] categories.
#         e.g. GSPO:.92000 GCAT:.88235 GDIP:.04347 (for k = 3).
function command_2 {
    stem=$1
    k=$2
    scores=""
    term=$(stem_to_term $stem)

    OLDIFS="$IFS"
    IFS=$','
    term_docs=(echo ${terms_dict[$term]})
    IFS="$OLDIFS"
    let a="${#term_docs[@]}"-1

    for m in ${!category_dict[@]}
    do
        OLDIFS="$IFS"
        IFS=$','
        category_docs=(echo ${category_dict[$m]})
        IFS="$OLDIFS"
        let b="${#category_docs[@]}"-1

        intersection=0
        for ((i=1; i <= ${#term_docs[@]} - 1; i++))
        do
            for ((j=1; j <= ${#category_docs[@]} - 1; j++))
            do
                if [[ ${term_docs[$i]} == ${category_docs[$j]} ]]
                then
                    intersection=`expr $intersection + 1`
                fi
            done
        done

        scores="$scores"$m":"$(jaccard_index "$a" "$b" "$intersection")"\n"
    done

    scores=$(echo -e $scores | sort -t":" -n -r +1)

    echo $(echo $scores | cut -f1-$k -d" ")
}

# params: $1: stem, $2: category
#
# term_docs: Array like this : [ ,doc_id_1, doc_id_2, doc_id_3, doc_id_4, doc_id_5, ...].
#            Contains documents that belong to a specific TERM (m).
#            NOTE: first element is " " and we ignore it (e.g. let a="${#term_docs[@]}"-1, watch -1)
# category_docs: Array like this : [ ,doc_id_1, doc_id_2, doc_id_3, doc_id_4, doc_id_5, ...].
#                Contains documents that belong to a specific CATEGORY ($1 argument).
#                NOTE: first element is " " and we ignore it (e.g. let b="${#category_docs[@]}"-1, watch -1)
# intersection: Essentially is: |category_docs INTERSECTION term_docs| (cardinality number).
# scores: Return value: string like this <JIScore(category, stem)> (e.g. .88235)
function command_3() {
    stem=$1
    category=$2
    term=$(stem_to_term $stem)

    OLDIFS="$IFS"
    IFS=$','
    term_docs=(echo ${terms_dict[$term]})
    category_docs=(echo ${category_dict[$category]})
    IFS="$OLDIFS"

    intersection=0
    for ((i=1; i <= ${#term_docs[@]} - 1; i++))
    do
        for ((j=1; j <= ${#category_docs[@]} - 1; j++))
        do
            if [[ ${term_docs[$i]} == ${category_docs[$j]} ]]
            then
                intersection=`expr $intersection + 1`
            fi
        done
    done

    let a="${#term_docs[@]}"-1
    let b="${#category_docs[@]}"-1
    let intersection="$intersection"
    echo $(jaccard_index "$a" "$b" "$intersection")
}

# params: $1: document, $2: option [-c|-t]
# OPTION: -t
#   term_temp: Variable that takes all the values of the
#   keys of the associative array terms_dict.
#
#   term_docs: Array like this : [ ,doc_id_1, doc_id_2, doc_id_3, doc_id_4, doc_id_5, ...].
#   Contains documents that belong to a specific TERM (term_temp).
#   NOTE: first element is " " and we ignore it
#   (e.g. for ((i=1; i <= ${#term_docs[@]} - 1; i++)), watch i=1 not 0)
#
#   If the document [$1] is in term_docs then we add the stem to the result (returning) string.
#
# OPTION: -c
#   category_temp: Variable that takes all the values of the
#   keys of the associative array category_dict.
#
#   category_docs: Array like this : [ ,doc_id_1, doc_id_2, doc_id_3, doc_id_4, doc_id_5, ...].
#   Contains documents that belong to a specific TERM (term_temp).
#   NOTE: first element is " " and we ignore it
#   (e.g. for ((i=1; i <= ${#category_docs[@]} - 1; i++)), watch i=1 not 0)
#
#   If the document [$1] is in category_docs then we add the category to the result (returning) string.
# OPTION: *else*
#   return "wrong option"
function command_4() {
    document=$1
    option=$2
    result=""

    OLDIFS="$IFS"
    IFS=$','
    if [[ "$option" == "-t" ]]
    then
        for term_temp in ${!terms_dict[@]}
        do
            term_docs=(echo ${terms_dict[$term_temp]})
            for ((i=1; i <= ${#term_docs[@]} - 1; i++))
            do
                temp="${term_docs["$i"]}"
                if [[ "$temp" == "$document" ]]
                then
                    result="$result"$(term_to_stem $term_temp)","
                    break
                fi
            done
        done
    elif [[ "$option" == "-c" ]]
    then
        for category_temp in ${!category_dict[@]}
        do
            category_docs=(echo ${category_dict[$category_temp]})
            for ((i=1; i <= ${#category_docs[@]} - 1; i++))
            do
                temp="${category_docs["$i"]}"
                if [[ "$temp" == "$document" ]]
                then
                    result="$result"$category_temp","
                    break
                fi
            done
        done
    else
        echo "wrong option"
    fi
    IFS="$OLDIFS"
    echo $result
}

# params: $1: document, $2: option [-c|-t]
#
# Essentially calls command_1 and counts how many
# commas [,] we got. That's the number of stems/categories as well.
function command_5() {
    OLDIFS="$IFS"
    IFS=$','
    echo $(wc -w <(command_4 $1 $2) | cut -f1 -d" ")
    IFS="$OLDIFS"
}

function cli() {
    read -r flag arg1 arg2

    case "$flag" in
    "@") echo $(command_1 "$arg1" "$arg2") ;;
    "#") echo $(command_2 "$arg1" "$arg2") ;;
    "$") echo $(command_3 "$arg1" "$arg2") ;;
    "P") echo $(command_4 "$arg1" "$arg2") ;;
    "C") echo $(command_5 "$arg1" "$arg2") ;;
    "quit") echo "Exiting..." ;;
    *) echo "Bad input. Try again." ;;
    esac
}

function main() {
    load_categories
    load_terms
    load_stems
    while true
    do
        echo -e "What do you want to do? \n1) Get the <k> most relevant stems for a <category>: @ <category> <k> \n2) Get the <k> most relevant categories for a <stem>: # <stem> <k> \n3) Get the Jaccard Index Score for a (category, stem) pair: $ <stem> <category> \n4) Get all the categories or stems of the document with <did> doc id (flag -c, -t): P <did> -c|-t \n5) Get the num of categories or num of stems of the document with <did> doc id (flag -c, -t): C <did> -c|-t \n6) Type quit to exit.\n\n"
        echo -n "Type here: "
        result=$(cli)
        if [[ "$result" == "Exiting..." ]]
        then
            break
        fi
        echo "Result: $result"
        echo -e "\n"
    done
    echo "Bye bye! :)"
}

main
````

##  Bash Script   News Sentiment Analyser

````python
%%file news_sentiment_analyser.sh
#!/bin/bash

# paths to the specific txt files
documents_path="/content/drive/MyDrive/Colab Notebooks/project2python/lyrl2004_vectors_train.dat.txt"
stems_path="/content/drive/MyDrive/Colab Notebooks/project2python/stem.termid.idf.map.txt"
positive_path="/content/drive/MyDrive/Colab Notebooks/project2python/positive_stems.txt"
negative_path="/content/drive/MyDrive/Colab Notebooks/project2python/negative_stems.txt"

# data structures for data
stems_list=()
positive_stems=()
negative_stems=()
stems_sentiment=()

# Loading the stems in an array [stems_list] where a stem's
# index plus 1 is the corresponding term_id.
#
# Loading only the necessary positive stems.
# Necessary, meaning:
#   1. they don't exist in negative stems
#   2. they exist in stems_list
# Accordingly, for negative stems.
function load_data() {
    stems_list=($(paste <(cut -f1 -d" " "$stems_path")))

    while IFS= read -r line; do
        if grep -qw "$line" "$negative_path"; then
            continue
        fi
        if grep -qw "$line" "$stems_path"; then
            positive_stems+=("$line")
        fi
    done < "$positive_path"

    while IFS= read -r line; do
        if grep -qw "$line" "$positive_path"; then
            continue
        fi
        if grep -qw "$line" "$stems_path"; then
            negative_stems+=("$line")
        fi
    done < "$negative_path"
}

# A function that for every stem in stems_list,
# maps the stem to a character [p(ositive)|n(egative)|
# u(nknown)] in the array stems_sentiment. Same with stems_list,
# term's sentiment = stems_sentiment[term's id - 1].
function mapping() {
    pos_ind=0
    neg_ind=0
    pos_temp=${positive_stems[$pos_ind]}
    neg_temp=${negative_stems[$neg_ind]}

    for ((stem_ind=0; stem_ind <= ${#stems_list[@]} - 1; stem_ind++))
    do
        stem_temp=${stems_list[$stem_ind]}
        if [[ "$stem_temp" == "$pos_temp" ]]
        then
            stems_sentiment[$stem_ind]="p"
            let pos_ind=$pos_ind+1
            pos_temp=${positive_stems[$pos_ind]}
        elif [[ "$stem_temp" == "$neg_temp" ]]
        then
            stems_sentiment[$stem_ind]="n"
            let neg_ind=$neg_ind+1
            neg_temp=${negative_stems[$neg_ind]}
        else
            stems_sentiment[$stem_ind]="u"
        fi
    done
}

# Main function that for every document, terms calculates and echoes its sentiment.
function main() {
    load_data
    mapping

    OLDIFS="$IFS"
    IFS=$'\n'
    file=$(head -100 $documents_path)
    for line in $file
    do
        IFS=$' '
        pos=0
        neg=0
        fields=($(echo $line))
        document=${fields[0]}
        for ((t=1; t <= ${#fields[@]} - 1; t++))
        do
            term=$(echo ${fields[$t]} | cut -f1 -d":")
            let index=$term-1
            case "${stems_sentiment[$index]}" in
            "p") let pos=$pos+1 ;;
            "n") let neg=$neg+1 ;;
            esac
        done

        if [[ $pos > $neg ]]
        then
            echo "Document $document has a positive sentiment."
        elif [[ $pos < $neg ]]
        then
            echo "Document $document has a negative sentiment."
        else
            echo "Document $document has a neutral sentiment."
        fi
    done
    IFS="$OLDIFS"
}

# redirecting main's output to the txt file
main > sentimentpernewsitem.txt
````

````python
%%file christmas_tree.py
import os, sys

def main():
    print(f"I am Parent[0] of all, my PID is {os.getpid()}")
    pid11 = os.fork()
    if pid11 == 0:
        print(f"I am Child[1.1] with pid: {os.getpid()} and my Parent id: {os.getppid()}")
        pid21 = os.fork()
        if pid21 == 0:
            print(f"I am Child[2.1] with pid: {os.getpid()} and my Parent id: {os.getppid()}")
            pid31 = os.fork()
            if pid31 == 0:
                print(f"I am Child[3.1] with pid: {os.getpid()} and my Parent id: {os.getppid()}")
                pid41 = os.fork()
                if pid41 == 0:
                    print(f"I am Child[4.1] with pid: {os.getpid()} and my Parent id: {os.getppid()}")
                else:
                    os.waitpid(pid41, os.WUNTRACED)
            else:
                os.waitpid(pid31, os.WUNTRACED)
        else:
            pid22 = os.fork()
            if pid22 == 0:
                print(f"I am Child[2.2] with pid: {os.getpid()} and my Parent id: {os.getppid()}")
                pid32 = os.fork()
                if pid32 == 0:
                    print(f"I am Child[3.2] with pid: {os.getpid()} and my Parent id: {os.getppid()}")
                    pid42 = os.fork()
                    if pid42 == 0:
                        print(f"I am Child[4.2] with pid: {os.getpid()} and my Parent id: {os.getppid()}")
                    else:
                        pid43 = os.fork()
                        if pid43 == 0:
                            print(f"I am Child[4.3] with pid: {os.getpid()} and my Parent id: {os.getppid()}")
                        else:
                            os.waitpid(pid42, os.WUNTRACED)
                            os.waitpid(pid43, os.WUNTRACED)
                else:
                    pid33 = os.fork()
                    if pid33 == 0:
                        print(f"I am Child[3.3] with pid: {os.getpid()} and my Parent id: {os.getppid()}")
                        pid44 = os.fork()
                        if pid44 == 0:
                            print(f"I am Child[4.4] with pid: {os.getpid()} and my Parent id: {os.getppid()}")
                        else:
                            os.waitpid(pid44, os.WUNTRACED)
                    else:
                        os.waitpid(pid32, os.WUNTRACED)
                        os.waitpid(pid33, os.WUNTRACED)
            else:
                os.waitpid(pid21, os.WUNTRACED)
                os.waitpid(pid22, os.WUNTRACED)
    else:
        pid12 = os.fork()
        if pid12 == 0:
            print(f"I am Child[1.2] with pid: {os.getpid()} and my Parent id: {os.getppid()}")
            pid23 = os.fork()
            if pid23 == 0:
                print(f"I am Child[2.3] with pid: {os.getpid()} and my Parent id: {os.getppid()}")
                pid34 = os.fork()
                if pid34 == 0:
                    print(f"I am Child[3.4] with pid: {os.getpid()} and my Parent id: {os.getppid()}")
                    pid45 = os.fork()
                    if pid45 == 0:
                        print(f"I am Child[4.5] with pid: {os.getpid()} and my Parent id: {os.getppid()}")
                    else:
                        os.waitpid(pid45, os.WUNTRACED)
                else:
                    os.waitpid(pid34, os.WUNTRACED)
            else:
                os.waitpid(pid23, os.WUNTRACED)
        else:
            os.waitpid(pid11, os.WUNTRACED)
            os.waitpid(pid12, os.WUNTRACED)
    return 0

if __name__ == "__main__":
    os._exit(main())
````

##   Mixed Piping

   code cell     three-way pipe .

````python
%%file mixed_piping.py
import os, sys, time

MESSAGES = ("I love PLH211          ", "The professor is boring", "but the curriculum is  ", "interesting & useful   ")
fifoname = "fifo_pipe"

def main():
    # anonymous pipe
    anon_pipe = os.pipe()

    # if anonymous pipe fails
    if not anon_pipe:
        print("Pipe call failed.")
        os.close(anon_pipe[0])
        os.close(anon_pipe[1])
        os._exit(1)

    # transformer process creation through fork
    transf_pid = os.fork()
    if transf_pid == -1: # fork failed
        print("Fork call failed.")
        os.close(anon_pipe)
        os._exit(1)
    elif transf_pid > 0: # only parent (source process) inside
        os.close(anon_pipe[0]) # ONLY write (don't need read)
        for message in MESSAGES:
            os.write(anon_pipe[1], message.encode())
        os.wait() # wait for your child (transformer process)
        os.close(anon_pipe[1]) # close write
    else: # only child (transformer process) inside
        os.close(anon_pipe[1]) # ONLY read (don't need write)

        if os.path.exists(fifoname): # remove named pipe if exists (it's a normal file!)
            os.remove(fifoname)
        os.mkfifo(fifoname) # create named pipe

        # output process creation through fork
        output_pid = os.fork()
        if output_pid == -1: # fork failed
            print("Fork call failed.")
            os.close(anon_pipe[0])
            os._exit(1)
        elif output_pid > 0: # only parent (transformer process) inside
            named_pipe = os.open(fifoname, os.O_WRONLY) # ONLY write
            for message in MESSAGES:
                read_message = os.read(anon_pipe[0], len(message))
                time.sleep(1) # for smooth output, redundant
                tokens = read_message.decode().split() # tokenize
                for t in tokens:
                    os.write(named_pipe, (t+"\n").encode())
            os.close(anon_pipe[0]) # close read
        else: # only child (output process) inside
            i = 0 # EOF counts, at 3 we are done expecting something...
            with open(fifoname, 'r') as pipein:
                while True:
                    line = pipein.readline()[:-1]
                    print(line)
                    if i == 3:
                        os._exit(0)
                    if not line:
                        i+=1
                        time.sleep(3) # give some time to write...
                        continue
    os._exit(0)

if __name__ == "__main__":
  os._exit(main())
````
